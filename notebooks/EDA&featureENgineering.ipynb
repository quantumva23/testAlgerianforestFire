### Algreian Forest Lives Dataset


```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv" , header=1)
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```


### Data Cleaning


```python
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["Region"]=df["Region"].astype(int)

```

```python
df.isnull().sum()
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.loc[122]
```

```python
df.columns
```

```python
# strip whitespace from column names (fixes names like " RH" or " Ws"), then convert the selected cols
df.columns = df.columns.str.strip()
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
objects=[features for features in df.columns if df[features].dtypes=='object']
objects
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
        
```

```python
df.info()
```

```python
df.describe()
```

```python
df.to_csv("Cleaned_forest_fires_dataset.csv",index=False)
```
